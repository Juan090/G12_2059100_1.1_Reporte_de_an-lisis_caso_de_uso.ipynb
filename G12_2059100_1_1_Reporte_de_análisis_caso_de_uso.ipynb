PROGRAMA DE CASO

La Comisión Federal de Electricidad tiene un sistema de tarificación que funciona de la siguiente manera:

1. Se tiene un Límite de Nivel de Tarifa (LNT), que generalmente anda alrededor de los 35 kilowatts por bimestre, pero se tiene que confirmar en cada emisión de recibos.

2. Cada bimestre, se realiza una lectura física a los medidores de cada casa, para registrar el consumo por bimestre de cada hogar.

3. Todo el consumo igual o menor a LNT, se considera consumo estándar.

4. Todo el consumo mayor que LNT se considera consumo alto volumen.

5. La tarifa por kilowatt de consumo estándar $45 pesos por kilowatt.

6. La tarifa por kilowatt para el consumo alto volumen es de $143 pesos por kilowatt.

7. Para el consumo estándar, el gobierno tiene un porcentaje de subsidio del 60%.

8. Para el consumo alto volumen, el gobierno tiene un porcentaje de subsidio del 25%.

9. El costo por consumo estándar es igual a multiplicar el consumo estándar por la tarifa por kilowatt de consumo estándar.

10. El costo por consumo alto volumen es igual a multiplicar el consumo alto volumen por la tarifa por kilowatt de consumo alto volumen.

11. El pago del gobierno es igual a sumar la multiplicación del costo de consumo estándar por el subsidio de consumo estándar, más la multiplicación del costo por consumo alto volumen por el subsidio alto volumen.

12. El pago del cliente es igual a sumar la multiplicación del costo de consumo estándar por 1 menos el subsidio de consumo estándar, más la multiplicación del costo por consumo alto volumen por 1 menos el subsidio alto volumen.

13.
Tanto la tarifa por kilowatt de consumo estándar (45.00), la tarifa por kilowatt de consumo alto volumen (143.00), el porcentaje de subsidio estándar (0.60), y el porcentaje de subsidio alto volumen (0.25) están definidas por la Ley de Energía, por lo cual son datos que se tienen.

14. Elaborar un programa que pregunte el consumo bimestral, en Kilowatts, así como el número de Kilowatts que represente el Límite de Nivel de Tarifa (LNT) del periodo. Deberá calcular cuánto pagará el gobierno por subsidios, y cuánto pagará el cliente.


###Ejemplo de ejecución 1:

Dame el consumo en Kilowatts del bimestre: 20

Dame la cantidad de Kilowatts que representa el Límite de Nivel de Tarifa: 35

El gobierno paga: $1,215.00

El cliente paga: $810.00

###Ejemplo de ejecución 2:

Dame el consumo en Kilowatts del bimestre: 45

Dame la cantidad de Kilowatts que representa el Límite de Nivel de Tarifa: 35

El gobierno paga: $1,572.50

El cliente paga: $1,882.50

| Dato | Tipo de Dato | Origen | Variable | Fórmula |
|----|----|----|----|----|
|Tarifa estándar|Float|Se tiene|tarifa_estandar|45.00|
|Tarifa alto volumen|Float|Se tiene|tarifa_alto_volumen | 143.00|
|Subsidio estándar|Float|Se tiene|subsidio_estandar|0.60|
|Subsidio alto volumen|Float|Se tiene|subsidio_alto_volumen|0.25|
|Límite de Nivel de Tarifa|Float|Se tiene|LNT|35.0|
|Consumo Bimestral|Float|Se pregunta|consumo_bimestral||
|Costo por Consumo Estándar|Float|Se calcula|costo_estandar|consumo_bimestral * tarifa_estandar|
|Costo por Consumo Alto Volumen|Float|Se calcula|costo_alto_volumen|(consumo_bimestral - LNT) * tarifa_alto_volumen if consumo_bimestral > LNT else 0|
|Pago del Gobierno|Float|Se calcula| pago_gobierno| (costo_estandar * subsidio_estandar) + (costo_alto_volumen * subsidio_alto_volumen)|
|Pago del Cliente|Float|Se calcula| pago_cliente|(costo_estandar * (1 - subsidio_estandar)) + (costo_alto_volumen * (1 - subsidio_alto_volumen))|

```python
#Programa de Caso
#Definición de variables
tarifa_estandar=45.00
tarifa_alto_volumen=143.00
subsidio_estandar=0.60
subsidio_alto_volumen=0.25
LNT=35.0

#Se pregunta el consumo bimestral en kilowatts y el LNT
consumo_bimestral=float(input("Ingrese el consumo bimestral en Kilowatts: "))
LNT=float(input("Ingrese el Límite de Nivel de Tarifa (LNT) del periodo en Kilowatts: "))

#Se calcula el costo por consumo estándar y alto volumen
costo_estandar=consumo_bimestral*tarifa_estandar
if consumo_bimestral>LNT:
    costo_alto_volumen=(consumo_bimestral-LNT)*tarifa_alto_volumen
else:
    costo_alto_volumen=0

#Se calcula el pago del gobierno y del cliente
pago_gobierno=(costo_estandar*subsidio_estandar)+(costo_alto_volumen*subsidio_alto_volumen)
pago_cliente=(costo_estandar*(1-subsidio_estandar))+(costo_alto_volumen*(1-subsidio_alto_volumen))

#Se muestra el resultado
print(f"El pago del gobierno por subsidios es: ${pago_gobierno:.2f}")
print(f"El pago del cliente es: ${pago_cliente:.2f}")
```
